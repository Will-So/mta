```python
import pandas as pd
from collections import defaultdict
from datetime import datetime

cols = ['ca', 'unit', 'scp', 'station', 'linename', 'division', 'date', 'time', 'desc', 'entries', 'exits']

traffic = pd.read_csv('data/all.csv', header = False, names = cols)

## group by first four columns, then add new column to 
## original dataframe that counts entries at this scp
## since the last measurement

traffic = traffic.loc[traffic.entries != 'ENTRIES']
traffic.entries = traffic.entries.astype(int)
traffic.exits = traffic.exits.astype(int)

traffic2 = traffic.groupby(['ca', 'unit', 'scp', 'station'])


traffic['EntryDeltas'] = traffic2.entries.diff()
## we are dropping the nan's that result from diff()
traffic = traffic.dropna(subset = ['EntryDeltas'])
traffic.head(90)
```

```python
small_traffic = traffic.sample(n=1e5)
```

```python
traffic['dt'] = pd.to_datetime(traffic.date + ' ' + traffic.time, format='%m/%d/%Y %H:%M:%S')
```

```python
traffic.date = pd.to_datetime(traffic.date)
```

## Removing Erroneous Data

```python
traffic = traffic[traffic.EntryDeltas < 5000]
traffic = traffic[traffic.EntryDeltas > 0]
```
